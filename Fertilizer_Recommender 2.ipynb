```python
X_train_np = X_train.values.astype(np.float32)
X_test_np = X_test.values.astype(np.float32)
y_train_np = y_train.values.astype(int)
y_test_np = y_test.values.astype(int)

print("X_train_np shape:", X_train_np.shape)
print("X_test_np shape:", X_test_np.shape)
print("y_train_np shape:", y_train_np.shape)
print("y_test_np shape:", y_test_np.shape)
```

## Define and train tabnet model

### Subtask:
Define the TabNet model architecture and train it on the training data.


**Reasoning**:
Import TabNetClassifier, instantiate the model, and train it on the prepared data.



```python
from pytorch_tabnet.tab_model import TabNetClassifier

# Instantiate TabNetClassifier
tabnet_clf = TabNetClassifier()

# Train model
tabnet_clf.fit(
    X_train_np, y_train_np,
    eval_set=[(X_test_np, y_test_np)],
    patience=10,
    max_epochs=100,
    virtual_batch_size=128
)
```

Use the trained TabNet model's encoder to extract features from the training and testing data and then print their shapes.



**Reasoning**:

## Train random forest model

### Subtask:
Train a Random Forest model using the original features combined with the features extracted from TabNet.


**Reasoning**:
Concatenate the original and TabNet features for both training and testing sets, initialize a RandomForestClassifier, and train the model.



```python
X_train_combined = np.concatenate((X_train.values, X_train_tabnet_features), axis=1)
X_test_combined = np.concatenate((X_test.values, X_test_tabnet_features), axis=1)

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_combined, y_train_np)

print("Random Forest model trained successfully.")
```

## Evaluate combined model

### Subtask:
